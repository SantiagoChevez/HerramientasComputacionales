```python
# numeric_columns = []
# for col in df.columns:
#     if df[col].dtype != "O":
#         numeric_columns.append(col)
numeric_columns = [col for col in df.columns if df[col].dtype != "O"]
object_columns = [col for col in df.columns if df[col].dtype == "O"]
```

```python
for col in numeric_columns:
    df[col].plot.box(title=col)
    plt.show()
```

```python
for col in numeric_columns:
    df[col].plot.kde(title=col)
    plt.show()
```

```python
print(f' Suma: {df["Year"].sum()}')
print(f' Promedio: {df["Year"].mean()}')
print(f' Mediana: {df["Year"].median()}')
print(f' Moda: {df["Name"].mode()}') # No se acostumbra con datos numéricosdf
print("Medidas de dispersión")
print(f' Desviación estándar: {df["Year"].std()}')
print(f' Varianza: {df["Year"].var()}')
print("Medidas de posición: ")
print(f' min: {df["Year"].min()}')
print(f' Q1: {df["Year"].quantile(0.25)}')
print(f' Q2: {df["Year"].quantile(0.50)}')
print(f' Q3: {df["Year"].quantile(0.75)}')
print(f' max: {df["Year"].max()}')
```

```python
#df[máscara]
df[df["Year"] <= 1990][["Name", "Platform"]]
```

```python
# Nombre de todos los juegos publicados por Nintendo después del 2000
# | -> or
# & -> and
# ~ -> not
df[(df["Year"] > 2000) & (df["Publisher"] == "Nintendo")]["Name"]
```

```python
# Mostrar los nombres de todos los juegos del psi/psx, n64 y saturn
#df["Platform"].unique() # Mustra todos los valored de la columna sin repetición
#df[(df["Platform"] == "PS") | (df["Platform"] == "PSX") | (df["Platform"] == "N64") | (df["Platform"] == "SAT")]["Name"]
df[df["Platform"].isin(["PS", "PSX", "N64", "SAT"])]["Name"]
```
