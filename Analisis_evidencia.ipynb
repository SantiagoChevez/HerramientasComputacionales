```python
import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('qs-world-rankings-2025_clean.csv')
```

```python
# Contar las ocurrencias de cada paÃ­s
conteo_paises = df["Location"].value_counts()

# Filtrar los paÃ­ses con un umbral mÃ­nimo de ocurrencias (por ejemplo, 10)
umbral = 5
paises_filtrados = conteo_paises[conteo_paises >= umbral]

# Crear la grÃ¡fica de pastel con los datos filtrados
paises_filtrados.plot.pie()
plt.title("PaÃ­ses con Mayor Presencia en el Ranking QS 2025")
plt.ylabel("")  # Opcional: para ocultar la etiqueta del eje y
plt.show()
```

```python
# Calcular el promedio de "QS Overall Score" por "Location"
promedios = df.groupby('Location Full')["QS Overall Score"].sum()

# Ordenar los promedios de mayor a menor y seleccionar los 10 primeros
top_10_promedios = promedios.sort_values(ascending=False).head(10)

# Crear la grÃ¡fica de barras con los 10 mejores promedios
top_10_promedios.plot.bar()
plt.title("Top 10 Paises en QS Overall Score por Location")
plt.ylabel("QS Overall Score")
plt.xlabel("Location")
plt.show()
```

```python
# Calcular el promedio de "QS Overall Score" por "Location"
promedios = df.groupby('Location Full')["Academic Reputation"].sum()

# Ordenar los promedios de mayor a menor y seleccionar los 10 primeros
top_10_promedios = promedios.sort_values(ascending=False).head(10)

# Crear la grÃ¡fica de barras con los 10 mejores promedios
top_10_promedios.plot.bar()
plt.title("Top 10 Paises de Academic Reputation Score por Location")
plt.ylabel("Suma Academic Reputation")
plt.xlabel("Location")
plt.show()
```
